All of the datasets in Anscombe's quartet have the same linear regression line, $y = 3.0 + 0.5x$ 

We can plot that line by executing the following:

```python
# generates 10 points from 3 to 14 and assigns the array to x
x = np.linspace(3,14,10)

# generates 10 points that satisfy (3 + 0.5*x) and assigns the array to y
y = 3 + 0.5*x

# makes a scatter plot with the x and y values
plt.scatter(x,y)

# adds a line plot to our scatter plot
plt.plot(x,y)

# shows the completed figure
plt.show()
```
